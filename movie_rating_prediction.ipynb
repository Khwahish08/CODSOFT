```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv('/content/imdb_movies_india.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Rating'])
```

```python
df = df.fillna({'Year': df['Year'].mode()[0], 'Duration': df['Duration'].mode()[0]})
df.head()
```

```python
df = df.fillna({'Votes' : df['Votes'].mode()[0]})
df.head()
```

```python
df.fillna({'Genre': 'Unknown',
           'Director': 'Unknown',
           'Actor1': 'Unknown',
           'Actor2': 'Unknown',
           'Actor3': 'Unknown'}, inplace=True)
```

```python
df.head()
```

```python
print(df['Year'].unique())
```

```python
df['Year'] = df['Year'].str.extract('(\d{4})')
df['Votes'] = df['Votes'].str.replace(r'[^\d]', '', regex=True)
df['Votes'] = df['Votes'].astype(int)
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(int)
df.head()
```

```python
df = pd.concat([df, df['Genre'].str.get_dummies(sep=', ')], axis=1)
df.head()
```

```python
df['Director'] = df['Director'].astype('category').cat.codes
df.head()
```

```python
df['Actor 1'] = df['Actor 1'].astype('category').cat.codes
df['Actor 2'] = df['Actor 2'].astype('category').cat.codes
df['Actor 3'] = df['Actor 3'].astype('category').cat.codes
df.head()
```

```python
X = df.drop(['Name', 'Rating', 'Genre'], axis=1)
X.head()
```

```python
X.shape
```

```python
y = df['Rating']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'RMSE: {rmse}, MAE: {mae}, R2: {r2}')
```
